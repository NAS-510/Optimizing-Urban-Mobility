```python
# Values of the attribute are in different scales. For example, holiday has maximum value 1 and traffic volume let say, 2000. Since these values are totally different range, it might be hurt model learning process. To overcome this issue, we need to do data normalization or standarization.
sc = MinMaxScaler()
sc_test = MinMaxScaler()
training_data_set = sc.fit_transform(train_data_without_ds)
training_data_frame = pd.DataFrame(training_data_set)
test_data_set = sc_test.fit_transform(test_data_without_ds)
test_data_frame = pd.DataFrame(test_data_set)
```

```python
# Make data for the LSTM model. I used past t-hours of data to predict the next 'n' hours data. Here window_size means window size data will be used for predicting next 'n' hours data

def df_to_x_y(data, window_size=5):
    X = []
    y = []
    for i in range(len(data)-window_size):
        row = [r for r in data[i:i+window_size]]
        X.append(row)
        label = data[i+window_size][num_of_feature-1]
        y.append(label)
    return np.array(X), np.array(y)

X, y = df_to_x_y(training_data_set,past_observation)
X.shape, y.shape
```

```python
# X test y Test
X_test, y_test = df_to_x_y(test_data_set, past_observation)
X_test.shape, y_test.shape
```

```python
# split data into train test and validation datase
from cgi import test


train_length = int(len(X)*0.7) # 90% trainging
validation_length = int(len(X)*0.3) # 10% testing and rest for validation
print(train_length, validation_length, (train_length+validation_length))
X_train, y_train = X[:train_length],y[:train_length]
X_val, y_val = X[train_length:], y[train_length:]
```

```python
(X_train.shape, y_train.shape), (X_val.shape, y_val.shape)
```

```python
# Define the LSTM_Model
LSTM_Model = Sequential()
# Add layers to the LSTM_Model
LSTM_Model.add(LSTM(hidden_layers_unit[0], input_shape=(past_observation, num_of_feature), return_sequences=return_sequence[0]))
LSTM_Model.add(Dropout(0.2))
LSTM_Model.add(LSTM(hidden_layers_unit[1], return_sequences=return_sequence[1]))
LSTM_Model.add(Dropout(0.2))
LSTM_Model.add(LSTM(hidden_layers_unit[2], return_sequences=return_sequence[2]))
LSTM_Model.add(Dropout(0.2))
LSTM_Model.add(LSTM(hidden_layers_unit[3], return_sequences=return_sequence[3]))
LSTM_Model.add(Dropout(0.2))
LSTM_Model.add(Dense(1))

# Compile the LSTM_Model
LSTM_Model.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.001), metrics=[tf.keras.metrics.MeanAbsoluteError(), tf.keras.metrics.MeanAbsolutePercentageError()])
```

```python
# Define the GRU_Model
GRU_Model = Sequential()
# Add layers to the GRU_Model
GRU_Model.add(GRU(hidden_layers_unit[0], input_shape=(past_observation, num_of_feature), return_sequences=return_sequence[0]))
GRU_Model.add(Dropout(0.2))
GRU_Model.add(GRU(hidden_layers_unit[1], return_sequences=return_sequence[1]))
GRU_Model.add(Dropout(0.2))
GRU_Model.add(GRU(hidden_layers_unit[2], return_sequences=return_sequence[2]))
GRU_Model.add(Dropout(0.2))
GRU_Model.add(GRU(hidden_layers_unit[3], return_sequences=return_sequence[3]))
GRU_Model.add(Dropout(0.2))
GRU_Model.add(Dense(1))

# Compile the GRU_Model
GRU_Model.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.001), metrics=[tf.keras.metrics.MeanAbsoluteError(), tf.keras.metrics.MeanAbsolutePercentageError()])
```

```python
LSTM_CheckPoint = ModelCheckpoint('/Users/nas/Documents/MRP/Traffic-Volume-Prediction-LSTM-GRU-main/model/model_lstm.keras', save_best_only=True)
GRU_CheckPoint = ModelCheckpoint('/Users/nas/Documents/MRP/Traffic-Volume-Prediction-LSTM-GRU-main/model/model_GRU.keras', save_best_only=True)
earlyStop =  tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=10, mode='min')
learning_rate_reduction = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.05, patience=3, min_lr=0.001)
```

```python
# Train LSTM Model
LSTM_history = LSTM_Model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=num_of_epoch, callbacks=[LSTM_CheckPoint,learning_rate_reduction], batch_size=batch_size, verbose=1)
```
